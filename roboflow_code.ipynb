```python
import cv2
import os
from ultralytics import YOLO
import random
import numpy as np
from collections import defaultdict
import csv
from google.colab.patches import cv2_imshow

# === CONFIGURATION ===
model_path = "/content/runs/detect/train/weights/best.pt"
input_path = "/content/VN20250520_120615 (2).mp4"
output_root ="/content/sample_data"
os.makedirs(output_root, exist_ok=True)

# === LOCATION SETUP ===
location_name = os.path.splitext(os.path.basename(input_path))[0]
location_folder = os.path.join(output_root, location_name)
os.makedirs(location_folder, exist_ok=True)

# === LOAD MODEL ===
model = YOLO(model_path)
names = model.names
colors = {i: (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)) for i in names}

# === TRACKING COUNTS ===
class_counts = defaultdict(int)

# === Resize and Pad Function ===
def resize_and_pad(image, target_width=1200, target_height=600):
    h, w = image.shape[:2]
    scale = min(target_width / w, target_height / h)
    new_w, new_h = int(w * scale), int(h * scale)
    resized = cv2.resize(image, (new_w, new_h))
    top = (target_height - new_h) // 2
    bottom = target_height - new_h - top
    left = (target_width - new_w) // 2
    right = target_width - new_w - left
    padded = cv2.copyMakeBorder(resized, top, bottom, left, right, cv2.BORDER_CONSTANT, value=(0, 0, 0))
    return padded

# === DETECTION FUNCTION ===
def detect_and_annotate(frame, track=False, confidence_threshold=0.05, iou_threshold=0.3, opacity=0.50):
    overlay = frame.copy()
    results = model.track(frame, persist=True, conf=confidence_threshold, iou=iou_threshold) if track else model.predict(frame, conf=confidence_threshold, iou=iou_threshold)

    if results[0].boxes:
        ids = results[0].boxes.id.cpu().numpy().astype(int) if results[0].boxes.id is not None else [None] * len(results[0].boxes.cls)
        boxes = results[0].boxes.xyxy.cpu().numpy().astype(int)
        class_ids = results[0].boxes.cls.int().cpu().tolist()

        for track_id, box, class_id in zip(ids, boxes, class_ids):
            x1, y1, x2, y2 = box
            label = names[class_id]
            color = colors.get(class_id, (255, 0, 0))
            class_counts[label] += 1
            label_text = f"{label}" if track_id is None else f"{label} ID:{track_id}"
            cv2.rectangle(overlay, (x1, y1), (x2, y2), color, -1)
            cv2.putText(overlay, label_text, (x1 + 3, y1 - 7), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)

    return cv2.addWeighted(overlay, opacity, frame, 1 - opacity, 0)

# === SAVE FRAME TO CATEGORY FOLDER ===
def save_frame_by_category(frame, frame_id):
    for label in class_counts:
        defect_folder = os.path.join(location_folder, label)
        os.makedirs(defect_folder, exist_ok=True)
        file_path = os.path.join(defect_folder, f"{label}_{frame_id}.jpg")
        cv2.imwrite(file_path, frame)

# === FILE TYPE DETECTION ===
file_ext = os.path.splitext(input_path)[1].lower()
is_image = file_ext in ['.jpg', '.jpeg', '.png']
is_video = file_ext in ['.mp4', '.avi', '.mov', '.mkv']

# === PROCESS IMAGE ===
if is_image:
    img = cv2.imread(input_path)
    if img is None:
        print("âŒ Failed to load image.")
        exit()
    annotated = detect_and_annotate(img, track=False)
    output_frame = resize_and_pad(annotated)
    save_frame_by_category(output_frame, "image")
    cv2_imshow(output_frame)

# === PROCESS VIDEO (1 frame every 5 seconds) ===
elif is_video:
    cap = cv2.VideoCapture(input_path)
    if not cap.isOpened():
        print("âŒ Failed to open video.")
        exit()

    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_interval = int(fps * 5)  # Every 5 seconds

    frame_count = 0
    saved_frame_id = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_count += 1
        if frame_count % frame_interval != 0:
            continue

        saved_frame_id += 1
        annotated = detect_and_annotate(frame, track=True)
        output_frame = resize_and_pad(annotated)
        save_frame_by_category(output_frame, saved_frame_id)
        cv2_imshow(output_frame)

    cap.release()

# === FINAL SUMMARY CSV ===
report_path = os.path.join(output_root, "final_defect_summary.csv")
report_data = []

for loc in os.listdir(output_root):
    loc_path = os.path.join(output_root, loc)
    if not os.path.isdir(loc_path) or loc == "final_defect_summary.csv":
        continue
    for category in os.listdir(loc_path):
        cat_path = os.path.join(loc_path, category)
        count = len([f for f in os.listdir(cat_path) if f.lower().endswith(('.jpg', '.png'))])
        report_data.append([loc, category, count])

with open(report_path, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Location', 'Defect Category', 'Image Count'])
    writer.writerows(report_data)

print(f"âœ… Final CSV report saved at: {report_path}")
```

```python
import os
import cv2
import pytesseract
from PIL import Image
import pandas as pd
from geopy.geocoders import Nominatim
from ultralytics import YOLO
from tqdm import tqdm

# === Install (if not already) ===
!apt install -y tesseract-ocr
!pip install pytesseract geopy ultralytics opencv-python --quiet

# === Config Paths ===
IMAGE_FOLDER = "/content/drive/MyDrive/final data/VN20250520_120615 (2)/Potholes"       # ðŸ” Change this
MODEL_PATH = "/content/drive/MyDrive/final data/runs/detect/train/weights/best.pt"                   # ðŸ” Change this
OUTPUT_CSV = "/content/defect_location_output.csv"
SAVE_BY_CATEGORY = True                           # âœ… Set True to save in folders

# === Tesseract path for Colab ===
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

# === Load YOLO model ===
model = YOLO("yolov8n.pt")

# === OCR Function with Bottom Crop ===
def extract_text_bottom(image_path):
    img = Image.open(image_path)
    width, height = img.size
    crop_height = int(height * 0.2)  # bottom 20%
    crop_box = (0, height - crop_height, width, height)
    cropped_img = img.crop(crop_box)
    return pytesseract.image_to_string(cropped_img)

# === Parse GPS coordinates ===
import re
def parse_gps(text):
    pattern = r"Lat[:\s]*([-+]?[0-9]*\.?[0-9]+)[,;\s]+Lon[:\s]*([-+]?[0-9]*\.?[0-9]+)"
    match = re.search(pattern, text, re.IGNORECASE)
    if match:
        lat = float(match.group(1))
        lon = float(match.group(2))
        return lat, lon
    return None, None

# === Reverse geocoding ===
def get_address(lat, lon):
    try:
        geolocator = Nominatim(user_agent="geoapi")
        location = geolocator.reverse((lat, lon), timeout=10)
        return location.address if location else "Unknown"
    except:
        return "Unknown"

# === Main Processing ===
def process_images():
    results = []

    for file in tqdm(os.listdir(IMAGE_FOLDER)):
        if not file.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue

        image_path = os.path.join(IMAGE_FOLDER, file)

        # 1. OCR for GPS
        text = extract_text_bottom(image_path)
        lat, lon = parse_gps(text)
        gps_address = get_address(lat, lon) if lat and lon else "Unknown"

        # 2. YOLO inference
        detection = model(image_path, verbose=False)[0]
        detected_classes = set()
        for box in detection.boxes.data.tolist():
            cls_id = int(box[5])
            detected_classes.add(model.names[cls_id])

        defect_label = ", ".join(detected_classes) if detected_classes else "None"

        # 3. Save frame to category folder (optional)
        if SAVE_BY_CATEGORY and defect_label != "None":
            for label in detected_classes:
                out_dir = os.path.join("/content/", label)
                os.makedirs(out_dir, exist_ok=True)
                out_path = os.path.join(out_dir, file)
                cv2.imwrite(out_path, cv2.imread(image_path))

        # 4. Save info
        results.append({
            "frame_name": file,
            "latitude": lat if lat else "NaN",
            "longitude": lon if lon else "NaN",
            "gps_address": gps_address,
            "defect_category": defect_label
        })

    # Save CSV
    df = pd.DataFrame(results)
    df.to_csv(OUTPUT_CSV, index=False)
    print(f"âœ… Output saved to {OUTPUT_CSV}")

process_images()
```

```python
def extract_metadata(image_path):
    results = reader.readtext(image_path)
    latitude = longitude = timestamp = "Unknown"

    for _, text, _ in results:
        text = text.strip().replace("Â°", "").replace("'", "").replace('"', '')
        lower_text = text.lower()
        if "lat" in lower_text:
            latitude = ''.join(filter(lambda x: x.isdigit() or x in ['.', '-'], text))
        elif "lon" in lower_text or "lng" in lower_text:
            longitude = ''.join(filter(lambda x: x.isdigit() or x in ['.', '-'], text))
        elif ":" in text and "-" in text:
            timestamp = text.strip()

    try:
        if latitude != "Unknown" and longitude != "Unknown":
            location = geolocator.reverse(f"{latitude}, {longitude}", language='en')
            address = location.address if location else "Address Not Found"
        else:
            address = "GPS Missing"
    except Exception as e:
        address = f"Geocoding Failed: {e}"

    return latitude, longitude, timestamp, address
```

```python
import os
import re
import csv
import cv2
import pytesseract
from PIL import Image, ImageEnhance

# === CONFIGURATION ===
image_folder = "/content/drive/MyDrive/final/images"
label_folder = "/content/drive/MyDrive/final/labels"
output_txt_folder = "/content/drive/MyDrive/final data"
output_csv_path = "/content/drive/MyDrive/final data/defect_summary_final.csv"

os.makedirs(output_txt_folder, exist_ok=True)

class_names = ['pothole', 'crack', 'speed_breaker', 'edge_line', 'lane_mark', 'edge_line']
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

# === OCR TEXT EXTRACTION ===
def extract_text(image_path):
    img = cv2.imread(image_path)
    h, w = img.shape[:2]
    cropped = img[int(h*0.85):, :]  # Bottom 15% of image

    gray = cv2.cvtColor(cropped, cv2.COLOR_BGR2GRAY)
    gray = cv2.resize(gray, None, fx=2, fy=2)
    pil_img = Image.fromarray(gray)
    pil_img = ImageEnhance.Contrast(pil_img).enhance(2)

    text = pytesseract.image_to_string(pil_img)
    return text

# === PARSE LOCATION AND GPS ===
def parse_location_and_gps(text):
    location, lat, lon = None, None, None

    match_dec = re.search(r'(\d{1,2}\.\d+)[,\s]+(\d{1,3}\.\d+)', text)
    if match_dec:
        lat = float(match_dec.group(1))
        lon = float(match_dec.group(2))

    if lat is None or lon is None:
        match_dms = re.search(
            r'(\d{1,2})\D+(\d{1,2})\D+(\d{1,2})?.*?[Nn].*?' +
            r'(\d{1,3})\D+(\d{1,2})\D+(\d{1,2})?.*?[Ee]', text)
        if match_dms:
            parts = match_dms.groups()
            d1, m1, s1 = int(parts[0]), int(parts[1]), int(parts[2] or 0)
            d2, m2, s2 = int(parts[3]), int(parts[4]), int(parts[5] or 0)
            lat = d1 + m1/60 + s1/3600
            lon = d2 + m2/60 + s2/3600

    for line in text.splitlines():
        if "Maharashtra" in line:
            location = line.strip()

    return location, lat, lon

# === READ LABELS ===
def read_labels(label_path):
    if not os.path.exists(label_path):
        return []
    with open(label_path, 'r') as f:
        lines = f.readlines()
        return list(set(class_names[int(line.split()[0])] for line in lines if line.strip()))

# === MAIN PROCESSING (TEXT EXTRACTION) ===
print("\nðŸ” Extracting text and metadata...")
for i, file in enumerate(os.listdir(image_folder)):
    if not file.lower().endswith('.jpg'):
        continue

    image_path = os.path.join(image_folder, file)
    label_path = os.path.join(label_folder, file.replace('.jpg', '.txt'))
    output_txt_path = os.path.join(output_txt_folder, file.replace('.jpg', '.txt'))

    ocr_text = extract_text(image_path)
    location, lat, lon = parse_location_and_gps(ocr_text)
    defects = read_labels(label_path)

    with open(output_txt_path, 'w', encoding='utf-8') as out:
        out.write(f"Frame: {file}\n")
        out.write(f"Location: {location or 'Unknown'}\n")
        out.write(f"Latitude: {lat if lat else 'NA'}\n")
        out.write(f"Longitude: {lon if lon else 'NA'}\n")
        out.write(f"Defects: {', '.join(defects) if defects else 'None'}\n")

    if i < 5:
        print(f"ðŸ–¼ï¸ Frame: {file}\nText:\n{ocr_text}\nParsed: Lat={lat}, Lon={lon}, Location={location}\n")

print("\nâœ… All frame metadata saved as .txt files.")

# === FINAL CSV CREATION ===
print("\nðŸ“¦ Creating final CSV...")
data = []
for fname in os.listdir(output_txt_folder):
    if not fname.endswith('.txt'):
        continue

    frame_name = fname.replace('.txt', '.jpg')
    with open(os.path.join(output_txt_folder, fname), 'r', encoding='utf-8') as f:
        lines = f.read()
        location_match = re.search(r'Location:\s*(.*)', lines)
        lat_match = re.search(r'Latitude:\s*(.*)', lines)
        lon_match = re.search(r'Longitude:\s*(.*)', lines)
        defect_match = re.search(r'Defects:\s*(.*)', lines)

        location = location_match.group(1) if location_match else 'Unknown'
        lat = lat_match.group(1) if lat_match else 'NA'
        lon = lon_match.group(1) if lon_match else 'NA'
        defects = defect_match.group(1).split(', ') if defect_match else []

        if lat != 'NA' and lon != 'NA' and defects:
            for defect in defects:
                data.append({
                    'frame': frame_name,
                    'label': defect,
                    'location': location,
                    'latitude': lat,
                    'longitude': lon
                })

# Save to CSV
with open(output_csv_path, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['frame', 'label', 'location', 'latitude', 'longitude']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)

print(f"\nâœ… Final CSV saved with {len(data)} entries at: {output_csv_path}")
```

```python
import os
import cv2
import pytesseract
import re
import csv
from PIL import Image, ImageEnhance, ImageFilter
from geopy.geocoders import Nominatim
from time import sleep

# === Config ===
image_folder = "/content/drive/MyDrive/final/images"
label_folder = "/content/drive/MyDrive/final/labels"
output_csv_path = "/content/defect_summary_final_fixed.csv"

# Classes from your model
class_names = ['pothole', 'crack', 'speed_breaker', 'edge_line', 'lane_mark', 'edge_line']

# Setup Tesseract
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

# Initialize geolocator
geolocator = Nominatim(user_agent="geoapi")

def extract_text_from_bottom(image_path):
    img = Image.open(image_path)
    w, h = img.size
    bottom_crop = img.crop((0, int(h * 0.85), w, h))
    bottom_crop = bottom_crop.convert("L")
    bottom_crop = ImageEnhance.Contrast(bottom_crop).enhance(2)
    return pytesseract.image_to_string(bottom_crop)

def extract_coordinates(text):
    text = text.replace("\n", " ").replace("Â°", ".")
    match = re.findall(r"(\d{2}\.\d{4,})", text)
    if len(match) >= 2:
        lat = float(match[0])
        lon = float(match[1])
        return lat, lon
    return None, None

def reverse_geocode(lat, lon, retries=3):
    for _ in range(retries):
        try:
            location = geolocator.reverse((lat, lon), timeout=10)
            if location:
                return location.address
        except Exception:
            sleep(1)
    return "Unknown"

def read_labels(label_path):
    if not os.path.exists(label_path):
        return []
    with open(label_path, 'r') as f:
        lines = f.readlines()
        return list(set(class_names[int(line.split()[0])] for line in lines if line.strip()))

# === Process All Images ===
data = []
for filename in os.listdir(image_folder):
    if not filename.endswith(".jpg"):
        continue

    img_path = os.path.join(image_folder, filename)
    label_path = os.path.join(label_folder, filename.replace('.jpg', '.txt'))

    text = extract_text_from_bottom(img_path)
    lat, lon = extract_coordinates(text)

    if lat and lon:
        location = reverse_geocode(lat, lon)
    else:
        location = "Unknown"

    labels = read_labels(label_path)

    for label in labels:
        data.append({
            'frame': filename,
            'label': label,
            'location': location,
            'latitude': lat if lat else 'NA',
            'longitude': lon if lon else 'NA'
        })

# === Save to CSV ===
with open(output_csv_path, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['frame', 'label', 'location', 'latitude', 'longitude'])
    writer.writeheader()
    writer.writerows(data)

print(f"âœ… Final CSV saved with {len(data)} entries at: {output_csv_path}")
```

```python
import os
import cv2

# === Set dataset path ===
base_dir = r"/content/drive/MyDrive/yolov12/road-anomalies-detection--5"
splits = ["train", "valid", "test"]
class_names = ['Edge line', 'Potholes', 'cracks', 'lane divider', 'objects', 'zebra crossing']

# === Define class colors ===
class_colors = {
    0: (0, 255, 0),       # Edge line â†’ Green
    1: (0, 0, 255),       # Potholes â†’ Red
    2: (255, 0, 0),       # Cracks â†’ Blue
    3: (0, 255, 255),     # Lane divider â†’ Yellow
    4: (255, 0, 255),     # Objects â†’ Magenta
    5: (255, 255, 0)      # Zebra crossing â†’ Cyan
}

def clean_label_file(label_path):
    cleaned_lines = []
    with open(label_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) != 5:
                print(f"âš ï¸ Removing bad line in {os.path.basename(label_path)}: {line.strip()}")
                continue
            cleaned_lines.append(" ".join(parts))
    with open(label_path, 'w') as f:
        for cl in cleaned_lines:
            f.write(cl + "\n")

for split in splits:
    image_dir = os.path.join(base_dir, split, "images")
    label_dir = os.path.join(base_dir, split, "labels")
    output_dir = os.path.join(base_dir, split, "labelled_images")
    os.makedirs(output_dir, exist_ok=True)

    for filename in os.listdir(label_dir):
        if not filename.endswith(".txt"):
            continue

        label_path = os.path.join(label_dir, filename)
        clean_label_file(label_path)

        image_file = filename.replace(".txt", ".jpg")
        image_path = os.path.join(image_dir, image_file)

        if not os.path.exists(image_path):
            print(f"âŒ Missing image for: {image_file}")
            continue

        image = cv2.imread(image_path)
        if image is None:
            print(f"âš ï¸ Could not read image: {image_file}")
            continue

        height, width, _ = image.shape

        with open(label_path, 'r') as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) != 5:
                    continue
                try:
                    cls_id, x_center, y_center, w, h = map(float, parts)
                    x1 = int((x_center - w / 2) * width)
                    y1 = int((y_center - h / 2) * height)
                    x2 = int((x_center + w / 2) * width)
                    y2 = int((y_center + h / 2) * height)

                    color = class_colors.get(int(cls_id), (200, 200, 200))  # Default gray if unknown
                    label = class_names[int(cls_id)] if int(cls_id) < len(class_names) else f"class_{int(cls_id)}"

                    cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)
                    cv2.putText(image, label, (x1, max(y1 - 8, 10)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
                except Exception as e:
                    print(f"âš ï¸ Drawing error in {filename}: {e}")

        output_path = os.path.join(output_dir, image_file)
        cv2.imwrite(output_path, image)

    print(f"âœ… Done labeling and cleaning: {split}")

print("\nðŸŽ‰ All images labeled with class-specific colors and label files cleaned successfully!")
```

```python

from ultralytics import YOLO
import cv2
import os
import glob

# === CONFIG ===
model_path = "/content/drive/MyDrive/final/runs/detect/train/weights/best.pt"  # Ensure this is the model exported from Roboflow
video_path = "/content/VN20250520_120615.mp4"  # Update path
frame_dir = "/content/frames"
feedback_dir = "/content/feedback_frames"
os.makedirs(frame_dir, exist_ok=True)
os.makedirs(feedback_dir, exist_ok=True)

# === EXTRACT FRAMES ===
cap = cv2.VideoCapture(video_path)
frame_list = []
frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frame_path = os.path.join(frame_dir, f"frame_{frame_count:05}.jpg")
    cv2.imwrite(frame_path, frame)
    frame_list.append(frame_path)
    frame_count += 1

cap.release()
print(f"âœ… Extracted {frame_count} frames.")

# === LOAD MODEL ===
model = YOLO(model_path)
print("âœ… Model loaded")

# === RUN DETECTION ===
no_detections = 0
low_conf_count = 0
CONFIDENCE_THRESHOLD = 0.5  # Match Roboflow setting

for frame_id, frame_path in enumerate(frame_list):
    img = cv2.imread(frame_path)

    # Resize if needed (match Roboflow input size, e.g. 640x640)
    img_resized = cv2.resize(img, (640, 640))

    results = model(img_resized)

    found = False
    for box in results[0].boxes:
        cls = int(box.cls.item())
        conf = float(box.conf.item())
        found = True

        print(f"âœ… Frame {frame_id}: Detected class {cls} with confidence {conf:.2f}")

        if conf < CONFIDENCE_THRESHOLD:
            low_conf_count += 1
            review_path = os.path.join(feedback_dir, f"frame_{frame_id:05}.jpg")
            cv2.imwrite(review_path, img)

            with open("/content/low_conf_log.csv", "a") as f:
                f.write(f"{frame_path},{cls},{conf:.2f}\n")

    if not found:
        no_detections += 1
        print(f"âš  Frame {frame_id}: No detections")

print(f"âœ… Detection complete: {no_detections} frames had no detections, {low_conf_count} low-confidence frames saved.")
```
